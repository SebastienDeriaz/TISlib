## imgLevelAdjust

```python
import TIS
import numpy as np
```

```python
f = np.array([[-128,-64,-32,0],[0,32,64,127]], dtype='int8')

TIS.showImage(f)

TIS.showHistogram(f)
```

```python
F = TIS.imgLevelAdjust(f, outDType='uint8', _mini=0, _maxi=100)

TIS.showHistogram(F)

print(F.dtype)
TIS.showImage(F)
```

## Paddings

```python
A = np.ones((10,10),dtype='uint8')
m = np.zeros((3,3))

A1 = TIS.imagePadding(A, m, 'zero')
A2 = TIS.imagePadding(A, m, 'replicate')
A3 = TIS.imagePadding(A, m, 'mirror')
A4 = TIS.imagePadding(A, m, 'circular')


print(A.dtype, "->", A1.dtype)
print(A.dtype, "->", A2.dtype)
print(A.dtype, "->", A3.dtype)
print(A.dtype, "->", A4.dtype)
```
