```python
import TIS
import numpy as np

f = np.array([[-128,-64,-32,0],[0,32,64,127]], dtype='int8')

TIS.showImage(f)

TIS.showHistogram(f)
```

```python
F = TIS.imgLevelAdjust(f, outDType='uint8', _mini=0, _maxi=100)

TIS.showHistogram(F)

print(F.dtype)
TIS.showImage(F)
```
